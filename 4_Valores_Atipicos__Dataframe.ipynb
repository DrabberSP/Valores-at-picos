```python
# Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Ventas_totales.csv')
```

```python
# Verificamos información del data frame
data.info()
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
# Reemplazamos valores nulos con bfill
data1=data.fillna(method='bfill')
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
# Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia
data2=data.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
# Corroboramos valores nulos del dataframe
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
# Realizamos diagrama de caja o bigote de cada una de las columnas
fig=plt.figure(figsize=(15,15))
data2.plot(kind='box',vert=False)
plt.title('Valores Atípicos del DataFrame')
plt.show()
```

Desviación estándar para eliminar Outliers en DataFrame

```python
# Método de desviación estándar
y=data2
Limite_Superior=y.mean()+ 3*y.std()
Limite_Inferior=y.mean()- 3*y.std()
print('Lista superior permitido', Limite_Superior)
print('Lista inferior permitido', Limite_Inferior)
```

```python
# Encontramos Outliers del DataFrame
outliers= data2[(y>Limite_Superior) | (y<Limite_Inferior)]
outliers
```

```python
# Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3=data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
# Eliminamos columnas innecesarias
data4=data3.drop(['indice_tiempo'],axis=1)
data4
```

```python
# Reemplazamos valores atipicos con "mean"
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
# Corroboramos valores nulos del DataFrame LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

Procedimiento Cuantiles para sustituir outliers en dataframe

```python
# Método aplicando cuantiles. Encuentro cuartiles 0.25 y 0.75
y=data2

percentile25=y.quantile(0.25)
percentile75=y.quantile(0.75)
iqr=percentile75 - percentile25

Limite_Superior_iqr=percentile75 + 1.5*iqr
Limite_Inferior_iqr=percentile25 - 1.5*iqr
print('Límite superior permitido', Limite_Superior_iqr)
print('Límite inferior permitido', Limite_Inferior_iqr)
```

```python
# Encontramos outliers del DataFrame
outliers_iqr=data2[(y>Limite_Superior_iqr) | (y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
# Obtenemos datos limpios del DataFrame
data3_iqr=data2[(y<=Limite_Superior_iqr) | (y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
# Eliminamos columnas innecesarias
data4_iqr=data3_iqr.drop(['indice_tiempo'],axis=1)
data4_iqr
```

```python
# Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
# Reemplazamos valores atipicos (nulos) del dataframe con "mean"
data5_iqr=data4_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data5_iqr
```

```python
# Unimos la columna
indice_tiempo=data2['indice_tiempo']
Datos_limpios=pd.concat([indice_tiempo, data5_iqr],axis=1)
Datos_limpios
```

```python
# Corroboramos valores nulos
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
# Convertir DataFrame a csv
Datos_limpios.to_csv('Datos_limpios.csv')
```
